```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import graphviz
```

```python
df = pd.read_csv('heart_disease.csv')
df.head(20)

```

```python
# show number of people with heart disease
df.target.value_counts()
```

```python
# show percentage of male and female.
count_female = len(df[df.sex == 0])
count_male = len(df[df.sex == 1])

print('Percentage of female patients: {: .2f}%'.format(count_female/(len(df.sex))*100))
print('Percentage of male patients: {: .2f}%'.format(count_male/(len(df.sex))*100))
```

```python
# show the mean
df.groupby('target').mean()
```

```python
# show NaN or null values
df.isnull().sum()
```

```python
# Train and test the model
# Create logistic regression
y = df.target.values
x = df.drop(['target'], axis=1)
```

```python
# split the data into testing and training sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# check shape of data after splitting
x_train.shape, x_test.shape
```

```python
# check the datatype
x_train.dtypes
```

```python
x_train.head()
```

```python

# Create instance of DecisonTreeClassifier
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
clf_gini.fit(x_train, y_train)
```

```python
# predict the y values using the decision tree classifier
y_pred_test = clf_gini.predict(x_test)

print(f'Accuracy score for the Gini Impurity classifier using test data: {accuracy_score(y_test, y_pred_test)}')
```

```python
y_pred_training = clf_gini.predict(x_train)
# View  predictions
y_pred_training
```

```python
# print the training and testing set of the accuracy score
print(f"Training set accuracy score: {clf_gini.score(x_train, y_train)}")
print(f"Testing set accuracy score: {clf_gini.score(x_test, y_test)}")
```

```python
# plot decision tree
plt.figure(figsize=[12,8])
tree.plot_tree(clf_gini.fit(x_train, y_train))
```

```python
# Create an instance of the DecisionTreeClassifier class from sklearn.tree
# The default criterion is 'Gini'
model = tree.DecisionTreeClassifier()
model.fit(x, y)
# get class names for dataset
class_names = df['target'].unique().astype(str)
#  Generate .dot file 
tree.export_graphviz(model, out_file='heart_disease_gini.dot', feature_names=x.columns, class_names=class_names, label='all', filled=True, rounded=True)
```
