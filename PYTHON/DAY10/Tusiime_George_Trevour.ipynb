```python
#importing all the necessary libraries
import pandas as pd
import matplotlib as plt 
import seaborn as sns
import numpy as np
```

```python
#reading the data sets
df=pd.read_csv('UceResultsBySchool2011-2016.csv')
df
```

```python
#removing white space from the column names
df = df.rename(columns=lambda x: x.strip())
print(df.columns)
```

```python
#checking for any null values in each column
null_values=df.isnull().sum()
print(null_values)
```

```python
#droping the null values and returning a dataset with no null values
df=df.dropna()
print(df)
```

```python
#VISUALIZING THE DATASET
"""
bar graphs
histogram
heatmap
line grap
pie chart
line plot

"""
```

```python


# Create a bar graph
plt.figure(figsize=(12, 6))
plt.bar(df['FEMALE TOTAL DIV1'], df['FEMALE CANDIDATES'])
plt.xlabel('Female Total DIV1')
plt.ylabel('Female Candidates')
plt.title('Female Candidates who got DIV1 in Different Years')
plt.tight_layout()
plt.show()
```

```python


# Get the unique districts
districts = df['DISTRICT'].unique()

# Count the number of occurrences for each district
district_counts = df['DISTRICT'].value_counts()

# Plot the pie chart
plt.figure(figsize=(100, 50))
plt.pie(district_counts, labels=districts, autopct='%1.1f%%')
plt.title('Distribution of Districts')
plt.axis('equal')
plt.show()
```

```python

grouped_data = df.groupby('YEAR')['TOTAL CANDIDATES'].sum()

# Plot the pie chart
plt.figure(figsize=(10, 10))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%')
plt.title('Total  Candidates by Year')
plt.axis('equal')
plt.show()
```
